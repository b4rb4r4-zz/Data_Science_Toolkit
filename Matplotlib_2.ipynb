# Tutorial Matplotlib

**Neste tutorial você verá como criar diferentes tipos de gráficos com o Matplotlib bem como alterar os estilos e  variar o mapa de cores com o colormap**

```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

## Scatter Plot

```python
x = np.linspace(0,50,100)
y = np.random.randint(0,50, size=(10,10))
```

```python
fig = plt.figure(figsize=(6,6))
plt.scatter(x,y)
```

### Alterando colormaps

```python
# A variável c cria um array de 0-100, que posteriormente será utilizada para variar os níveis de cor
c = np.arange(100) 
```

**Os diferentes tipos de colormaps podem ser encontrados na documentação oficial do [Matplotlib](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html).**
Abaixo alguns exemplos:

```python
cmaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
cmaps_2 = ['Pastel1', 'Pastel2', 'Paired', 'Accent',
            'Dark2', 'Set1', 'Set2', 'Set3',
            'tab10', 'tab20', 'tab20b', 'tab20c']
```

```python
fig = plt.figure(figsize=(10,6))
plt.scatter(x,y, c=c, cmap=cmaps_2[6])
plt.colorbar()
```

### Alterando outras propriedades do scatter plot

```python
fig = plt.figure(figsize=(10,6))
plt.scatter(x,y, marker='*', alpha=1, linewidths=2,edgecolors='red')
```

### Alterando estilo

```python
plt.style.use('seaborn-white')
fig = plt.figure(figsize=(10,6))
plt.scatter(x,y, marker='*', alpha=1, linewidths=2,edgecolors='red')
```

## Histograma

```python
from random import sample
data = sample(range(1, 1000), 100)
fig = plt.figure(figsize=(8,6))
plt.hist(data, bins=20)
```

### Alterando outras propriedades do Histograma

Confira a documentação do [Matplotlib](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.hist.html) para ver mais customizações do hstograma.

```python
counts, bins = np.histogram(data)
plt.hist(bins[:-1], bins, weights=counts, histtype='bar', orientation='horizontal')
```

## Boxplot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# box plot retangular
plt.boxplot(data,vert=True,patch_artist=True);  
```
