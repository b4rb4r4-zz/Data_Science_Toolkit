### Factorplot

```python
tips = sns.load_dataset('tips')
tips.head()
```

**O factorplot é a forma mais genérica de plotar gráficos da relação entre uma ou mais variáveis númericas com dados categóricos. Aceita o parâmetro ``kind`` para mudar o tipo de gráfico. As opções são: ``point``, ``bar``, ``strip``, ``swarm``, ``box``, ``violin``, or ``boxen``**

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='boxen')
```

### Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Podemos orientar os dados para aparecerem na horizontal
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

### Violinplot

Um violinplot desempenha um papel semelhante a um boxplot. Ele mostra a distribuição de dados quantitativos em vários níveis de uma (ou mais) variáveis categóricas, de modo que essas distribuições possam ser comparadas. Ao contrário de um boxplot, no qual todos os componentes do gráfico correspondem a pontos de dados reais, o gráfico de violino possui uma estimativa da densidade do núcleo da distribuição subjacente.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

### Stripplot

O stripplot irá desenhar um scatterplot onde uma variável é categórica. Um stripplot pode ser desenhado por conta própria, mas também é um bom complemento para uma boxplot ou violinplot nos casos em que você deseja mostrar todas as observações juntamente com alguma representação da distribuição subjacente.

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

### Swarmplot

O swarmplot é semelhante ao stripplot (), mas os pontos são ajustados (somente ao longo do eixo categórico) para que eles não se sobreponham. Isso dá uma melhor representação da distribuição de valores, embora não se ajude também a um grande número de observações (tanto em termos de capacidade de mostrar todos os pontos quanto em termos da computação necessária para organizá-los).

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

### Barplot

Esses plots parecidos permitem que você obtenha dados agregados de um recurso categórico. ** barplot ** é um gráfico geral que permite que você agregue os dados categóricos baseados em alguma função, por padrão, a média:

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

### Countplot

Isto é essencialmente o mesmo que o gráfico de barras, exceto que o estimador está explicitamente contando o número de ocorrências. É por isso que apenas passamos o valor x:

```python
sns.countplot(x='sex',data=tips)
```

### Configurando as palettes

Para utilizar uma paleta nativa do Seaborn, você pode utilizar o comando ``sns.color_palette()``. A partir desse comando, você também pode criar a sua própria paleta inserindo as cores no formato de tuplas RGB, códigos de cores hexadecimal, ou utilizar os nomes das cores no padrão HTML.

As paletas nativas do Seaborn são: ``deep``, ``muted``,``pastel``, ``dark``, ``bright``e ``colorblind``.

```python
bright = sns.color_palette("bright")
deep = sns.color_palette("deep")
muted = sns.color_palette("muted")
pastel = sns.color_palette("pastel")
colorblind = sns.color_palette("colorblind")
dark = sns.color_palette("dark")

sns.palplot(deep)
sns.palplot(muted)
sns.palplot(pastel)
sns.palplot(dark)
sns.palplot(bright)
sns.palplot(colorblind)
```

**Para utilizar a paleta que você escolheu, utilize o comando ``sns.set_palette()``**

```python
sns.set_palette("pastel")

# Plot utilizando a paleta pastel
sns.stripplot(x="day", y="total_bill", data=tips)
```

### Paleta de Cores Brewer

Esta paleta de cores é inspirada no cartógrafo Cindy Brewer. Ela está dividida em três categorias: ``Diverging``, ``Qualitative`` e ``Sequential``.

- As paletas de cores **Qualitative** possuem cores distintas e recomenda-se utilizá-la quando não houver uma relação sequencial entre os dados categóricos.   


- As paletas de cores **Sequential** possuem uma escala degradê de cores que facilita a visualização de dados com uma sequência específica.   


- As paletas de cores **Diverging** devem ser utilizadas quando deseja-se evidenciar dados extremos, como por exemplo, em mapas de calor, em que deseja-se visualizar com mais ênfase as temperaturas mais frias ou mais quantes.

Para mais detalhes, consulte a documentação do [Seaborn](https://seaborn.pydata.org/tutorial/color_palettes.html?highlight=color).


**Exemplo de paleta de cores Qualitative**

```python
qualitative_colors = sns.color_palette("Set3", 10)
sns.palplot(qualitative_colors)
```

**Exemplo de paleta de cores Sequential**

```python
sequential_colors = sns.color_palette("RdPu", 10)
sns.palplot(sequential_colors)
```

**Exemplo de paleta e cores Diverging**

```python
diverging_colors = sns.color_palette("RdBu", 10)
sns.palplot(diverging_colors)
```

### Customizando a sua própria paleta

- Utilizando código [hexadecimal](https://www.color-hex.com/)
- Utilizando [nome das cores](https://python-graph-gallery.com/100-calling-a-color-with-seaborn/)

```python
my_colors = ['#000080','#3b5998', '#007f7f', '#afeeee','#e6e6fa', '#f55b70','#ff6767','#cc3333','#852332']

sns.palplot(sns.color_palette(my_colors))
```

```python
my_colors = ['black', 'grey', 'salmon', 'b', 'cyan', 'navy']

sns.palplot(sns.color_palette(my_colors))
```
