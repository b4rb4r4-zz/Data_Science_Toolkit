# Tutorial da biblioteca Matplotlib

Neste tutorial será apresentado os principais gráficos que podem ser desenvolvidos a partir da biblioteca Matplotlib. Os gráficos mais utilizados em Data Sicence são gráficos de barras, boxplots, scatter plots e histogramas.

Você aprenderá como criar um gráfico, configurar o nome dos eixos, título, o tipo de linhas, alterar as cores das linhas, as dimensões dos gráficos, inserir legendas, exibir subplots e alterar o estilo do plot.

Em caso de dúvida, consulte a documentação do [Matplotlib](https://matplotlib.org/3.1.0/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py).

## Criando gráficos de linha do zero

```python
from matplotlib import pyplot as plt

%matplotlib inline 
```

O comando **%matplotlib inline** é utilizado para que os gráficos sejam plotados diretamente no output do jupyter sem que haja necessidade de inserir o comando **plt.show()**.

### A forma mais simples de criar gráficos no matplotlib é utilizando os comandos abaixo:

```python
import numpy as np

x = np.arange(0,100)
plt.plot(x, x**2, 'r') # 'r' é a cor vermelha (red)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
plt.show()
```

```python
# plt.subplot(nlinhas, ncols, numeroplot)
plt.subplot(1,2,1)
plt.plot(x, x**2, 'r--') # Mais nas seções de cores abaixo
plt.subplot(1,2,2)
plt.plot(x, x**3, 'g*-');
```

### Orientação a objetos

No entanto, para tornar os gráficos mais personalizáveis, a documentação do Matplotlib recomenda a abordagem orientada a objetos.

```python
fig = plt.figure() # Criando uma figura vazia sem eixos.
ax = fig.add_axes([0,0,1,1]) # Adiciona eixos à figura [esquerda, inferior, largura, latura] valores de 0-1

# Plota nos eixos especificados
ax.plot(x, x**2, 'b')
ax.set_xlabel('Define o Label X') # Perceba o uso de "set_" para começar métodos
ax.set_ylabel('Define o Label Y')
ax.set_title('Define o Título')
```

Em uma única figura podemos adicionar mais de um eixo

```python
fig = plt.figure() 
ax1 = fig.add_axes([0,0,1,1])             # eixo principal
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eixo secundário

# Plota o eixo principal
ax1.plot(x, x**2, 'b')
ax.set_xlabel('Define o Label X') 
ax.set_ylabel('Define o Label Y')
ax.set_title('Principal')

# Plota o eixo secundário
ax2.plot(x, x, 'r')
ax2.set_xlabel('Define o Label X') 
ax2.set_ylabel('Define o Label Y')
ax2.set_title('Secundário')
```

### Subplots

```python
fig, ax = plt.subplots(ncols= 2, nrows=1)
```

```python
ax
```

A variável que armazena os eixos é uma matriz, portanto, é utilizada a indexação para acessar os gráficos separadamente.

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(8,8)) #figsize altera o tamanho da figura

ax[0].plot(x, x**2, x, x**3)
ax[0].set_title("Graph 1")

ax[1].plot(x, x**2, x, x**3)
ax[1].set_title("Graph 2")
```

**Limitando eixos X e Y**

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(8,6))

ax[0].plot(x, x**2, x, x**3)
ax[0].set_title("Graph 1")

# limitando os eixos x e y
ax[1].plot(x, x**2, x, x**3)
ax[1].set_ylim([0, 60])
ax[1].set_xlim([2, 5])
ax[1].set_title("Graph 2")
```

**Legendas**

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(8,6))

ax[0].plot(x, x**2, label='Função Quadrática')
ax[0].plot(x, x**3, label = 'Função Cúbica')
ax[0].set_title("Graph 1")
ax[0].legend()

# limitando os eixos x e y
ax[1].plot(x, x**2, label = 'Quadrática')
ax[1].plot(x, x**3, label = 'Zoom')
ax[1].set_ylim([0, 60])
ax[1].set_xlim([2, 5])
ax[1].set_title("Graph 2")
ax[1].legend()
```

As legendas podem ser mudadas do posição de acordo com o valor de `loc`

```python
ax[1].legend(loc=1) # Canto superior direito
ax[1].legend(loc=2) # Canto superior esquerdo
ax[1].legend(loc=3) # Canto inferior esquerdo
ax[1].legend(loc=4) # Canto inferior direito

ax[1].legend(loc=0) # deixe matplotlib decidir a localização ideal
fig
```

## Alterando cores, estilos e marcadores

### Forma simplificada

Na forma simplificada, as cores e marcadores são atribuída a uma string que é inserida opcionalmente no método plot() após a inserção dos valores de x e y.
 A lista completa com todas as abreviações para cores e marcadores encontram-se na documentação oficial do [Matplotlib](https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html).

```python
fig, ax = plt.subplots(1,2,figsize=(16,6))
ax[0].plot(x, x**3, 'b--') # linha azul tracejada
ax[1].plot(x, x**2, 'g>') # linha verde com marcador triangular
```

```python
fig, ax = plt.subplots(1,2,figsize=(16,6))
ax[0].plot(x, x**3, linewidth= 4.5) # linewidth ou lw é a espessura da linha
ax[1].plot(x, x**2, 'g>', alpha=0.5) # alpha está entre 0-1 e representa a transparencia da linha
```
